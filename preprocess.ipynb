```python
from pathlib import Path
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import shutil
from tqdm import tqdm_notebook as tqdm
%config InlineBackend.figure_formats = {'png', 'retina'}
```

```python
img_path = Path("./celeba/img_align_celeba/")
img_path = img_path.glob("*.jpg")
```

```python
image_list = []
for path in tqdm(img_path):
    arr = plt.imread(path)
    w, h, c = arr.shape
    if w < 128 or h < 128 or c !=3:
        continue
    image_list.append(path)
```

```python
len(image_list)
```

```python
prob = [0.8,0.1,0.1]
dir = ["train","val","test"]
base_path = Path("./celeba/img_align_celeba/")

for path in tqdm(image_list):
    dst_path = base_path.parent / Path(np.random.choice(dir, p=prob))
    
    dst_path.mkdir(exist_ok=True)
    shutil.copy(path, dst_path / path.name)
```
